```python
data['streams'] = pd.to_numeric(data['streams'], errors = 'coerce')
data.dtypes
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
print("success!")
```

```python
# Graph 1

top_artists = data['artist(s)_name'].value_counts().head(10)

plt.figure(figsize=(10,5))
plt.barh(top_artists.index, top_artists.values, color='navy')


plt.xlabel("Number of Songs")
plt.ylabel("Artist(s) Name")
plt.title("Top 10 Artists with Most Songs")
plt.show()

print(top_artists)
```

```python
# Graph 2

top_spotify_streams = data[['track_name', 'artist(s)_name', 'streams']].sort_values(by='streams', ascending = False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_spotify_streams['streams'], y=top_spotify_streams['track_name'], palette='viridis')
plt.xlabel("Streams (billions)")
plt.ylabel("Track Name")
plt.title("Top 10 Songs with most Streams on Spotify")
plt.show()

top_spotify_streams
```

```python
# Graph 3

plt.figure(figsize=(10,6))
plt.hist(data['danceability_%'], bins=20, color='navy', alpha=0.7, rwidth = 0.9)
plt.xlabel('Danceability (%)')
plt.ylabel('Frequency')
plt.title('Distribution of Danceability')
plt.grid(True)
# Show the plot
plt.show()
```

```python
# Graph 4

# Calculating average danceability for each year
average_danceability_by_year = data.groupby('released_year')['danceability_%'].mean()

# Plotting trends in danceability over the years
plt.figure(figsize=(10, 6))
plt.plot(average_danceability_by_year.index, average_danceability_by_year.values, color='blue', marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Average Danceability (%)')
plt.title('Trends in Danceability Over the Years')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Heatmap 1

# columns for cross-platform metrics
cross_platform_columns = [
    'in_spotify_playlists', 'in_spotify_charts', 'in_apple_playlists', 
    'in_apple_charts', 'in_deezer_playlists', 'in_deezer_charts', 'in_shazam_charts'
]

# correlation matrix
correlation_matrix = data[cross_platform_columns].corr()

# Plotting the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap: Cross-Platform Metrics')
plt.show()
```

```python
# Graph 5

audio_features_columns = ['danceability_%', 'energy_%', 'valence_%']
popularity_column = 'streams'

# Creating scatter plots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for idx, feature in enumerate(audio_features_columns):
    axes[idx].scatter(data[feature], data[popularity_column], color='blue', alpha=0.5)
    axes[idx].set_xlabel(feature.replace('_', ' ').title())
    axes[idx].set_ylabel('Number of Streams')
    axes[idx].set_title(f'{feature.replace("_", " ").title()} vs. Number of Streams')

plt.tight_layout()
plt.show()
```

```python
# Graph 6

# columns for audio features and popularity
audio_features_columns = ['danceability_%', 'energy_%', 'valence_%']
popularity_column = 'streams'

# Creating scatter plots
plt.figure(figsize=(12, 4))
for feature in audio_features_columns:
    plt.scatter(data[feature], data[popularity_column], alpha=0.5, label=feature.replace('_', ' ').title())

plt.xlabel('Audio Feature')
plt.ylabel('Number of Streams')
plt.title('Relationship Between Audio Features and Song Popularity')
plt.legend()
plt.show()
```

```python
# Graph 7

# Creating scatter plot for Popularity vs. Danceability
plt.figure(figsize=(10, 4))
plt.scatter(data['danceability_%'], data['streams'], color='blue', alpha=0.5)
plt.xlabel('Danceability (%)')
plt.ylabel('Number of Streams')
plt.title('Popularity vs. Danceability')
plt.show()
```

```python
# Graph 8

# Creating box plot for Valence distribution by Year
plt.figure(figsize=(10, 4))
data.boxplot(column='valence_%', by='released_year', grid=False)
plt.xlabel('Year')
plt.ylabel('Valence (%)')
plt.title('Valence Distribution by Year')
plt.suptitle('')
plt.show()
```

```python
# Heatmap 2


# Select columns for audio features
audio_features_columns = ['danceability_%', 'valence_%', 'energy_%', 'acousticness_%', 'instrumentalness_%']

# Calculate the correlation matrix
correlation_matrix = data[audio_features_columns].corr()

# Create a heatmap
plt.figure(figsize=(10, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap: Audio Features')
plt.show()
```
